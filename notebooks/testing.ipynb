```python
import pandas as pd
import sys
sys.path.append("../src")
from eda import EDASuite

# Load dataset
df = pd.read_csv("../data/sample.csv")

# Initialize EDA suite
eda = EDASuite(df)

# Summary stats
print("Summary Statistics:\n", eda.summary_stats())

# Missing values
print("\nMissing Values:\n", eda.missing_values())

# Correlation
print("\nCorrelation Matrix:\n", eda.correlations())

# Generate plots
eda.plot_histograms()
eda.plot_correlation_heatmap()
print("\nPlots saved in 'plots/' folder")
```
